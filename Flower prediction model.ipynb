```python
# Removing outliers
```

```python
Q1 = np.percentile(iris_df['sepal width (cm)'], 25)
Q3 = np.percentile(iris_df['sepal width (cm)'], 75)
IQR = Q3 - Q1

# Calculate the lower and upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound,upper_bound)
```

```python
# iris_df['sepal width (cm)']=np.where(iris_df['sepal width (cm)']>upper_bound
outliers=np.where((iris_df['sepal width (cm)'] >= upper_bound) | (iris_df['sepal width (cm)'] <= lower_bound))
outliers
# outliers=iris_df[outliers]
# outliers
iris_df.loc[15]
```

```python
print(iris_df.dtypes)
```

```python
print(upper_bound.dtype)
```

```python
iris_df['sepal width (cm)']=np.where((iris_df['sepal width (cm)'] >= upper_bound),
                                     upper_bound,
                                     np.where((iris_df['sepal width (cm)'] <= lower_bound),
                                              lower_bound,
                                              iris_df['sepal width (cm)']))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(iris_df.iloc[:,0:3],iris_df.iloc[:,4],test_size=0.2,random_state=42)
xtrain.shape
```

```python
iris_df.iloc[:,4]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
y_pred=lr.predict(xtest)
```

```python
accuracy_score(ytest,y_pred)
```

```python
x=iris_df.iloc[:,0:3]
y=iris_df.iloc[:,4]
```

```python
from sklearn.model_selection import cross_val_score
```
